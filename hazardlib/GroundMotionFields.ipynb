# Ground motion fields calculation in openquake.hazardlib

```python
from openquake.hazardlib.source import Rupture
from openquake.hazardlib.geo import Point, Line, SimpleFaultSurface, RectangularMesh
from openquake.hazardlib.calc import ground_motion_fields
from openquake.hazardlib.gsim.campbell_bozorgnia_2008 import CampbellBozorgnia2008
from openquake.hazardlib.site import Site, SiteCollection
from openquake.hazardlib.imt import PGA
from openquake.hazardlib.correlation import JB2009CorrelationModel

import numpy
from matplotlib import pyplot
from mpl_toolkits.basemap import Basemap
```

```python
def get_grid_and_map_proj(surf, buf=0.3, delta=0.01):
    """
    Return grid of nodes and map projection specific to surface
    """
    min_lon, max_lon, max_lat, min_lat = surf.get_bounding_box()
    
    min_lon -= buf
    max_lon += buf
    min_lat -= buf
    max_lat += buf

    lons = numpy.arange(min_lon, max_lon + delta, delta)
    lats = numpy.arange(min_lat, max_lat + delta, delta)
    lons, lats = numpy.meshgrid(lons, lats)
    mesh = RectangularMesh(lons=lons, lats=lats, depths=None)
    
    m = Basemap(projection='merc', llcrnrlat=numpy.min(lats), urcrnrlat=numpy.max(lats),
                llcrnrlon=numpy.min(lons), urcrnrlon=numpy.max(lons), resolution='l')
    
    return mesh, m
```

# Ground motion fields from simple fault rupture

```python
# create Simple Fault Surface
surface = SimpleFaultSurface.from_fault_data(
    fault_trace=Line([Point(9.21602706445, 45.1555287905), Point(9.25645636929, 45.1877167851),
                  Point(9.29688464252, 45.2199047798), Point(9.35715705075, 45.2398017764),
                  Point(9.42902686305, 45.2401237764), Point(9.47246500782, 45.2381597767),
                  Point(9.51590215304, 45.236194777), Point(9.56736930079, 45.2307927779),
                  Point(9.61883544823, 45.2253897788), Point(9.67030259419, 45.2199877797),
                  Point(9.72270625188, 45.2033947825), Point(9.77510990175, 45.1868007853),
                  Point(9.83238881096, 45.1680237884), Point(9.88966771001, 45.1492457915),
                  Point(9.94694559775, 45.1304687947), Point(10.0042244753, 45.1116907978)]),
    upper_seismogenic_depth=2.,
    lower_seismogenic_depth=7.,
    dip=30.,
    mesh_spacing=2.
)

# create rupture
rupture = Rupture(
    mag=7.,
    rake=0.,
    tectonic_region_type='Active Shallow Crust',
    hypocenter=surface.get_middle_point(),
    surface=surface,
    source_typology=None
)
```

```python
# define sites where to compute ground motion fields
mesh, m = get_grid_and_map_proj(surface)
sites = SiteCollection([Site(location=Point(lon, lat), vs30=760., vs30measured=True, z1pt0=40., z2pt5=1.)
                        for lon, lat in zip(mesh.lons.flatten(), mesh.lats.flatten())])
```

```python
# compute median ground motion field by setting truncation level = 0
fields = ground_motion_fields(
    rupture=rupture,
    sites=sites,
    imts=[PGA()],
    gsim=CampbellBozorgnia2008(),
    truncation_level=0,
    realizations=1
)
```

```python
# plot median ground motion field
x, y = m(mesh.lons, mesh.lats)

fig = pyplot.figure(figsize=(9, 9))
m.contourf(x, y, fields[PGA()].reshape(mesh.shape), 100)
m.drawparallels(numpy.arange(numpy.min(mesh.lats), numpy.max(mesh.lats), 0.2), labels=[True, False, False, True])
m.drawmeridians(numpy.arange(numpy.min(mesh.lons), numpy.max(mesh.lons), 0.2), labels=[True, False, False, True])
m.colorbar()
pyplot.title('Median Ground Motion Field', fontsize=20)
```

```python
# compute ground motion field including aleatory variability up to three sigmas (truncation level = 3)
numpy.random.seed(123)
fields = ground_motion_fields(
    rupture=rupture,
    sites=sites,
    imts=[PGA()],
    gsim=CampbellBozorgnia2008(),
    truncation_level=3,
    realizations=2
)
```

```python
for i in range(fields[PGA()].shape[1]):
    f = fields[PGA()][:, i]
    
    fig = pyplot.figure(figsize=(9, 9))
    m.contourf(x, y, f.reshape(mesh.shape), 100)
    m.drawparallels(numpy.arange(numpy.min(mesh.lats), numpy.max(mesh.lats), 0.2), labels=[True, False, False, True])
    m.drawmeridians(numpy.arange(numpy.min(mesh.lons), numpy.max(mesh.lons), 0.2), labels=[True, False, False, True])
    m.colorbar()
    pyplot.title('Ground Motion Field Realization %s' % (i + 1), fontsize=20)
```

```python
# compute ground motion field including aleatory variability up to three sigmas (truncation level = 3)
# and spatial correlation
numpy.random.seed(123)
fields = ground_motion_fields(
    rupture=rupture,
    sites=sites,
    imts=[PGA()],
    gsim=CampbellBozorgnia2008(),
    truncation_level=3,
    realizations=2,
    correlation_model=JB2009CorrelationModel(vs30_clustering=False)
)
```

```python
for i in range(fields[PGA()].shape[1]):
    f = array(fields[PGA()])[:, i]
    
    fig = pyplot.figure(figsize=(9, 9))
    m.contourf(x, y, f.reshape(mesh.shape), 100)
    m.drawparallels(numpy.arange(numpy.min(mesh.lats), numpy.max(mesh.lats), 0.2), labels=[True, False, False, True])
    m.drawmeridians(numpy.arange(numpy.min(mesh.lons), numpy.max(mesh.lons), 0.2), labels=[True, False, False, True])
    m.colorbar()
    pyplot.title('Ground Motion Field Realization %s with spatial correlation' % (i + 1), fontsize=20)
```
