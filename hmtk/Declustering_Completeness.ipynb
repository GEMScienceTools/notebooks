### Import the required libraries

### Import the catalogue

### Declustering Example 1: Gardner & Knopoff (1974) using the original Gardner & Knopoff Distance and Time Windows

#### Set-up the tool

```python
# Set up the declustering algorithm
# Step 1 - set-up the tool
gardner_knopoff = GardnerKnopoffType1()
```

The Gardner & Knopoff algorithm requires two configurable inputs:

1) "time_distance_window" 
   Different methods for determining the magnitude scaling of the time and distance windows are selected

2) "fs_time_prop"
   This implementation of the algorithm searches through the catalogues in descending order of magnitude, the size of the time-window used to search for 
   foreshocks can be represented as a proportion of the aftershock window. Thus:

  i) Do not consider foreshocks (fs_time_prop = 0.0)
 ii) Foreshock window < Aftershock window (fs_time_prop < 1.0)
iii) Foreshock time window equal to aftershock window (fs_time_prop = 1.0)
 iv) Foreshock time window greater than aftershock window (fs_time_prop > 1.0)

#### Create the configuration file

```python
declust_config = {'time_distance_window': GardnerKnopoffWindow(),
                  'fs_time_prop': 1.0}
print declust_config
```

#### Run the declustering

```python
print 'Running declustering ...'
vcl, flag_vector = gardner_knopoff.decluster(catalogue, declust_config)
print 'done!'
print '%s clusters found' % np.max(vcl)
print '%s Non-poissionian events identified' % np.sum(flag_vector != 0)
```

```python
data = np.column_stack([catalogue.get_decimal_time(), catalogue.data['magnitude'], catalogue.data['longitude'], catalogue.data['latitude'], vcl, flag_vector])
print '      Time    Magnitude    Long.    Lat.   Cluster No. Index'
for row in data:
    print '%14.8f  %6.2f  %8.3f  %8.3f  %6.0f  %6.0f' %(row[0], row[1], row[2], row[3], row[4], row[5])
```

## Declustering Example 2: AFTERAN (Musson, 1999) using the Gardner & Knopoff Distance and Time Windows

```python
# Setup the algorithm
afteran = Afteran()
declust_config = {'time_distance_window': GardnerKnopoffWindow(),
                  'time_window': 60.}
print 'Running Afteran ...'
# Run Afteran
vcl2, flag_vector2 = afteran.decluster(catalogue, declust_config)
print 'done!'
print '%s clusters found' % np.max(vcl2)
print '%s Non-poissionian events identified' % np.sum(flag_vector2 != 0)
```

### Purging the Catalogue of Non-Poissonian events

```python
catalogue.select_catalogue_events(flag_vector == 0)
print 'Purged catalogue now contains %s events' % catalogue.get_number_events()
```

## Completeness Analysis Example: Stepp (1971) Methodology

```python
The Stepp (1971) methodology was originally developed as a graphical method for estimating completeness.
The duration for which a particular bin is complete is measured by the duration over which the magnitudes
follow a Poisson process. The catalogue will be considered not complete for magnitudes in a particular bin
when the observed rate drops below the expected Poisson rate.

To adapt from a graphical to a computation method the deviation point for each bin is calculated by 
optimising a piecewise linear fit to the expected rate (with three constrained optimisation parameter)
and the cross-over point is used to mark the end of the complete period. Results can be verified graphically.
```

#### Set-up algorithm

The configuration file contains three variables:

1) 'magnitude_bin': The size of the magnitude bin (float)

2) 'time_bin': The size of the time-step to increment the time analysis (float)

3) 'increment_lock': A boolean ("True"/"False") variable to determine whether the trend in the completeness magnitude
    should be fixed ("True") in order to ensure the trend of lower completeness magnitudes for the later part of the
    catalogue, or to allow the completeness to vary without constraint ("False").

```python
stepp = Stepp1971()
#
completeness_config = {'magnitude_bin': 0.5,
                       'time_bin': 5.0,
                       'increment_lock': False}
print completeness_config
```

#### Run the completeness analysis

```python
# Run analysis
print 'Running Stepp (1971) completeness analysis:'
completeness_table = stepp.completeness(catalogue, completeness_config)
print 'done!'

# Print the output completeness table
for row in completeness_table:
    print '%8.1f  %8.2f' %(row[0], row[1])
```
